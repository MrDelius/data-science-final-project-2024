```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ndtv_data_final.csv')
```

```python
df.head()
```

```python
# first check column names
df.columns
```

```python
# then check types
df.dtypes
```

```python
# now lets move categoriacal variables
# first i am going to work with continuous variables
cat_feat = ['Unnamed: 0', 'Name', 'Brand', 'Model', 'Touchscreen', 'Operating system', 'Wi-Fi', 'Bluetooth', 'GPS', '3G', '4G/ LTE']
data = df.drop(cat_feat, axis=1, inplace=False)
data.head()
```

```python
# let's see basical data
data.describe()
```

```python
data.groupby('Processor').mean()
```

```python
data.isnull().sum()
```

```python
# lets make normalization method and take those columns that can effect price

# with the method of max and min normalization we can make values interactive
normalized_df = data.apply(lambda x: (x - x.min())/ (x.max()-x.min()))

normalized_df
```

```python
# Now lets calculate correlation of numeric columns to price
data.corr()
# Plotting correlation heatmap
dataplot = sns.heatmap(data.corr(numeric_only=True), cmap="YlGnBu", annot=True)

# Displaying heatmap
plt.show()
```

```python
# i want to exclude number of sims becase i has negative correlation
normalized_df.drop('Number of SIMs', axis=1, inplace=True)
if "Number of SIMs" in normalized_df.columns:
    print("Not removed!")
else:
    print("Removed!")
```

```python
# to make the priorities of columns that have stronger effect to price we will multiply correlation numbers to the column values
# calculate correlation of each column with the target (e.g., Price)
correlations = data.corr()["Price"].drop(["Price", "Number of SIMs"])  # Exclude self-correlation
correlations.index
```

```python
# do not forget that data variable is normalized to normalized_df variable, but correlation is checked with actual data
print(normalized_df.head())
effective_data = normalized_df[correlations.index].mul(correlations.values, axis=1)
effective_data.head()
```

```python
effective_data["Quality"] = effective_data.mean(axis=1)
effective_data["Price"] = normalized_df["Price"]
effective_data["Value"] = np.where(
    (effective_data["Price"] != 0) & (effective_data["Quality"] != 0),  
    effective_data["Price"] / effective_data["Quality"],
    pd.NA # Value if False
)
effective_data.head()
```

```python
client_profit = effective_data['Value'].min()
client_profit = effective_data[effective_data['Value']==client_profit]
client_profit
```

```python
business_profit = effective_data['Value'].max()
business_profit = effective_data[effective_data['Value']==business_profit]
business_profit
```

```python
effective_data["Price"].max()
```

```python
t = effective_data[['Quality', 'Value']]
t.loc[:,'Quality'] = t['Quality']*data['Price']
t.loc[:,'Value'] = t['Value']*data['Price']
final_data = pd.concat([data, t], axis=1)
final_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
```

```python
X = data.drop(columns=['Price'])
y = data['Price']  # Target variable (price)

# normalizer = MinMaxScaler()
# normed_data = normalizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

# linear regression 
lr = LinearRegression()
lr.fit(X_train, y_train)

# as i understand the model tries to predict price for new brand of mobile phones by given new features (unseen X_test values)
y_pred_linear = lr.predict(X_test)

# now we compare predicted values with y_test which is actual value
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
score_tst_ln = lr.score(X_test, y_test)
score_trn_ln = lr.score(X_train, y_train)

print("Linear Regression Evaluation:")
print("MSE:", mse_linear)
print("R-squared:", r2_linear)
print("R2 of test data:", score_tst_ln)
print("R2 of train data:", score_trn_ln)

# polynomial regression
poly = PolynomialFeatures(degree=2)
# first we learn data and then transform learned data
X_poly_train = poly.fit_transform(X_train)
# transform already learned data
X_poly_test = poly.transform(X_test)

# use linear regression because it is fast and not complex. it thinks of pow(x, 4)=n4
poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)

y_pred_poly = poly_model.predict(X_poly_test)

# model Evaluation
# Polynomial Regression Evaluation
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)
score_tst_pl = poly_model.score(X_poly_test, y_test)
score_trn_pl = poly_model.score(X_poly_train, y_train)

print("\nPolynomial Regression Evaluation:")
print("MSE:", mse_poly)
print("R-squared:", r2_poly)
print("R2 of test data:", score_tst_pl)
print("R2 of train data:", score_trn_pl)
```

```python
# сalculate residuals for both models
residuals_linear = y_test - y_pred_linear
residuals_poly = y_test - y_pred_poly

# plotting distribution of residuals
plt.figure(figsize=(12, 6))

# Linear Regression Residuals Distribution
plt.subplot(1, 2, 1)
sns.histplot(residuals_linear, kde=True, color='blue', bins=20)
plt.title('Linear Regression: Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')

# Polynomial Regression Residuals Distribution
plt.subplot(1, 2, 2)
sns.histplot(residuals_poly, kde=True, color='green', bins=20)
plt.title('Polynomial Regression: Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import Ridge
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_poly_train, y_train)

y_train_ridge = ridge_model.predict(X_poly_train)
y_test_ridge = ridge_model.predict(X_poly_test)

train_mse = mean_squared_error(y_train, y_train_ridge)
test_mse = mean_squared_error(y_test, y_test_ridge)
train_r2 = r2_score(y_train, y_train_ridge)
test_r2 = r2_score(y_test, y_test_ridge)

print("Train MSE:", train_mse)
print("Train R2:", train_r2)
print("Test MSE:", test_mse)
print("Test R2:", test_r2)
```

```python
residuals_test = y_test - y_test_ridge
residuals_train = y_train - y_train_ridge

# plotting distribution of residuals
plt.figure(figsize=(12, 6))

# Linear Regression Residuals Distribution
plt.subplot(1, 2, 1)
sns.histplot(residuals_test, kde=True, color='blue', bins=20)
plt.title('Test for Ridge: Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')

# Polynomial Regression Residuals Distribution
plt.subplot(1, 2, 2)
sns.histplot(residuals_train, kde=True, color='green', bins=20)
plt.title('Train for Ridge: Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=5, weights='uniform')  # weights can be 'uniform' or 'distance'
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

mse_knn = mean_squared_error(y_test, y_pred)
r2_knn = r2_score(y_test, y_pred)

print("k-NN Regression Evaluation:")
print("MSE:", mse_knn)
print("R-squared:", r2_knn)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize the Decision Tree Regressor
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model
dt_model.fit(X_train, y_train)

# Make predictions
y_pred_dt = dt_model.predict(X_test)

# Evaluate the model
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)

print("Decision Tree Regression Evaluation:")
print("MSE:", mse_dt)
print("R-squared:", r2_dt)
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid
param_grid = {
    'max_depth': [6, 7, 8, 9, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2'],
    'max_leaf_nodes': [None, 10, 20],
    'criterion': ['mse', 'friedman_mse']
}

grid_search = GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit grid search
grid_search.fit(X_train, y_train)

# Best parameters and best model
print("Best Parameters:", grid_search.best_params_)
best_dt_model = grid_search.best_estimator_

# Evaluate the tuned model
y_pred_best = best_dt_model.predict(X_test)
mse_best = mean_squared_error(y_test, y_pred_best)
r2_best = r2_score(y_test, y_pred_best)

print("Tuned Decision Tree Regression Evaluation:")
print("MSE:", mse_best)
print("R-squared:", r2_best)
```
